```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
from scipy.integrate import odeint
from scipy.integrate import solve_ivp
```

# First Order ODEs
Differential equation
$$ \frac{dv}{dt} - \alpha v^2 + \beta = 0 \hspace{10mm} v(0) = 0 $$
or, $$ \frac{dv}{dt} = \alpha v^2 - \beta $$

```python
## fuction to define ode
def dvdt(t, v):
    return 4*v**2 - 6
v0 = 0
```

```python
t = np.linspace(0, 1, 100)  #timespan for solving the equation
```

```python
t
```

```python
sol1 = odeint(dvdt, y0=v0, t=t, tfirst=True) # solving using odient solver
```

```python
sol1.T
```

```python
plt.plot(t, sol1)
```

```python
sol2 = solve_ivp(dvdt, t_span=(0, max(t)), y0=[v0], t_eval=t)
```

```python
sol2.y
```

```python
# plotting the solution
plt.plot(t, sol2.y[0])
```

```python
# plotting two different solutions together
plt.plot(t, sol1.T[0])
plt.plot(t, sol2.y[0],'--')
plt.xlabel('$t$', fontsize=22)
plt.ylabel('$v(t)$', fontsize=22)
plt.show()
```
